```python
import matplotlib.pyplot as plt
import numpy as np

import KGCN


batch_arguments = [
    # varied random seeds
    '--dataset product --random_seed 6666',
    '--dataset product --random_seed 59',
    '--dataset product --random_seed 123',
    
    # # varied mixers and aggregators
    # '--dataset product --mixer transe --n_epochs 150 --random_seed 114',
    # '--dataset product --mixer attention --n_epochs 150 --random_seed 114',
    # '--dataset product --aggregator concat --mixer transe --n_epochs 150 --random_seed 114',
    # '--dataset product --aggregator concat --mixer attention --n_epochs 150 --random_seed 114',
    # '--dataset product --aggregator neighbor --mixer transe --n_epochs 150 --random_seed 114',
    # '--dataset product --aggregator neighbor --mixer attention --n_epochs 150 --random_seed 114',
]
```

```python
def plot_result(result, arg):
    """
    :param result: A tuple, (loss_list, test_loss_list, auc_score_list)
    :param arg: A str, hyper parameters
    """
    fig = plt.figure()
    gs = fig.add_gridspec(2, hspace=0)
    ax1, ax2 = gs.subplots(sharex=True)
    fig.suptitle(arg, size='medium', weight='regular', x=0.1, y=0.97, ha='left')
    
    ax1.plot(result[0], label='Train Loss')
    ax1.plot(result[1], label='Evaluation Loss')
    ax1.legend()
    ax2.plot(result[2], color='g', label='ROC AUC')

    ax1.set(ylabel='Loss')
    ax1.grid(linestyle = '--', linewidth = 0.5)
    ax2.set(xlabel='Epoch', ylabel='ROC AUC')
    ax2.grid(linestyle = '--', linewidth = 0.5)
    
    plt.tight_layout()
    
    plt.show()

def print_result_summary(result, arg):
    train = np.array(result[0])
    test = np.array(result[1])
    auc = np.array(result[2])

    i = np.argmin(train)
    j = np.argmin(test)
    k = np.argmax(auc)
    print('[Result] (epoch, train loss, test loss, auc)')
    print(f'|     Parameters: {arg}')
    print(f'│ Min train loss: {i:>3}, {train[i]:2.6f}, {test[i]:2.6f}, {auc[i]:2.6f}')
    print(f'│  Min test loss: {j:>3}, {train[j]:2.6f}, {test[j]:2.6f}, {auc[j]:2.6f}')
    print(f'└    MAX ROC AUC: {k:>3}, {train[k]:2.6f}, {test[k]:2.6f}, {auc[k]:2.6f}')
```

```python
# Show result
for arg in batch_arguments:
    print(f'[Info] Experiment: {arg}')
    result = KGCN.experiment(arg)
    plot_result(result, arg)
    print_result_summary(result, arg)
    print()

    # Remember to check the convergence speed, performance at convergence, best AUC
```
